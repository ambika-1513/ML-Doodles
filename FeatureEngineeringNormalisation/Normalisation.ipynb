```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('wine_data.csv', header = None , usecols = [0,1,2])
```

```python
df
```

```python
df.columns= ['Class label', 'Alchol', 'Malic acid']
```

```python
df
```

```python
sns.kdeplot(df['Alchol'])
```

```python
sns.kdeplot(df['Malic acid'])
```

```python
color_dict = {1:'yellow', 3:'pink', 2:'green'}
sns.scatterplot( x = 'Alchol', y = 'Malic acid', hue = 'Class label' ,data = df, palette = color_dict)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Class label' , axis =1), df['Class label'] , test_size = 0.33 , random_state = 15)
```

```python
X_train.shape , X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
sc.fit(X_train)
x_tr_sc = sc.transform(X_train)
x_te_sc  =sc.transform(X_test)
```

```python
X_train.describe()
```

```python
x_tr_sc.describe()
```

```python
x_tr_sc = pd.DataFrame(x_tr_sc , columns = X_train.columns)
x_te_sc = pd.DataFrame(x_te_sc , columns = X_test.columns)
```

```python
x_tr_sc.describe()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

ax1.scatter(X_train['Alchol'], X_train['Malic acid'],c=y_train)
ax1.set_title("Before Scaling")
ax2.scatter(x_tr_sc['Alchol'], x_tr_sc['Malic acid'],c=y_train)
ax2.set_title("After Scaling")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Alchol'], ax=ax1)
sns.kdeplot(X_train['Malic acid'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(x_tr_sc['Alchol'], ax=ax2)
sns.kdeplot(x_tr_sc['Malic acid'], ax=ax2)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['Alchol'], ax=ax1)


# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(x_tr_sc['Alchol'], ax=ax2)

plt.show()
```
